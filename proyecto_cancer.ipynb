```python
df_continentes = pd.read_csv('./files/dataset-continentes.csv')
df_continentes  
```

```python
df_continentes.isnull().sum()  
```

```python
# Borramos la ultima fila, que es la que contiene los nulos
df_continentes = df_continentes.iloc[:-1]
df_continentes
```

```python
# Borramos las columnas que no necesitamos
df_continentes.drop(columns=['Alpha-3 code', 'Cancer code', 'Population code (ISO/UN)', 'Country', 'Sex', '95% UI low', '95% UI high', 'Number.1'], inplace=True)
```

```python
df_continentes
```

```python
# Cambiamos el nombre de algunas columnas para que sea mÃ¡s intuitivo
df_continentes.rename(columns={"Label": "Continent", "Number": "Total Cases", "ASR (World)": "Age-Standardized Rate"}, inplace=True)
df_continentes
```

```python
# Guardamos el archivo limpio
df_continentes.to_csv('./clean files/dataset-continentes-limpio.csv', index=False)
```
